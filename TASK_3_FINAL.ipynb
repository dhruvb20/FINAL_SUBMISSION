```python
import pandas as pd
```

```python
df1=pd.read_csv("C:/Users/user/Downloads/userGroup_1.csv")
```

```python
df1_temp=df1[df1['key']=="previous_first_open_count"]
```

```python
df1_newusers=df1_temp[df1_temp['int_value']==0]
```

# Consider the case of addition of songs by newusers and there retentivity

```python
newusers=list(df1_newusers['user_pseudo_id'].unique())
```

```python
stickiness=[]
songs_added_count=[]
for i in newusers:
    df_temp=df1[df1['user_pseudo_id']==i]
    df_temp_2=df_temp[df_temp['event_name']=="number_of_items_event"]
    df_temp_3=df_temp_2[df_temp_2['key']=="item_count"]
    stickiness.append(len(df_temp['event_date'].unique()))
    songs_added_count.append(len(df_temp_3))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(stickiness)
plt.show()
```

```python
plt.hist(songs_added_count)
plt.show()
```

```python
plt.scatter(songs_added_count,stickiness)
plt.xlabel("songs_added_count")
plt.ylabel("stickiness_of_users")
plt.show()
```

# Consider the case of addition for 7D and 10D active users

```python
users=list(df1['user_pseudo_id'].unique())
```

```python
stickiness_7D=[]
songs_added_count_7D=[]
stickiness_10D=[]
songs_added_count_10D=[]
count=0
for i in users:
    print(count)
    count=count+1
    df_temp=df1[df1['user_pseudo_id']==i]
    df_temp_2=df_temp[df_temp['event_name']=="number_of_items_event"]
    df_temp_3=df_temp_2[df_temp_2['key']=="item_count"]
    if(len(df_temp['event_date'].unique())>=7):
        stickiness_7D.append(df_temp['event_date'].unique())
        songs_added_count_7D.append(len(df_temp_3))
    if(len(df_temp['event_date'].unique())>10):
        stickiness_10D.append(df_temp['event_date'].unique())
        songs_added_count_10D.append(len(df_temp_3))
```

```python
stick_7D=[]
stick_10D=[]
for i in stickiness_7D:
    stick_7D.append(len(i))
for i in stickiness_10D:
    stick_10D.append(len(i))
```

```python
plt.hist(stick_10D)
plt.show()
```

```python
plt.scatter(songs_added_count_7D,stick_7D)
plt.xlabel("songs_added_count")
plt.ylabel("stickiness_of_users")
plt.show()
```

```python
plt.scatter(songs_added_count_10D,stick_10D)
plt.xlabel("songs_added_count")
plt.ylabel("stickiness_of_users")
plt.show()
```

# Analysis
On analyzing I found out that for new users adding a new song leads to an increase in retentivity and considering 7D and 10D active users I found out that for these users adding new songs does not have much effect on their retentivity as observed from scatter and histogram plots.

A continuous variation of retentivity and the addition of new songs could be seen for the 7D and 10D active users.
